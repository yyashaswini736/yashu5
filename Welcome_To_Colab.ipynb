```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
from sklearn.metrics import mean_squared_error,mean_absolute_error
df=pd.read_csv('/content/BostonHousing (1).csv')
print(df)
df.info()
print(df.isnull().sum())
print(df.head())
plt.scatter(df['rm'],df.medv,color='Red',marker='*')
plt.xlabel('rm')
plt.ylabel('medv')
plt.title('Scatterplot')
plt.show()
x=df.drop('medv',axis=1)
# Fill missing values in the 'rm' column with the mean
x['rm'] = x['rm'].fillna(x['rm'].mean())
print(x)
y=df.medv
print(y)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
print("x_train shape:",x_train.shape)
print("x_test shape:",x_test.shape)
print("y_train shape:",y_train.shape)
print("y_test shape:",y_test.shape)
reg=linear_model.LinearRegression()
reg.fit(x,y)
y_pred=reg.predict(x_test)
print(reg.coef_)
print(reg.intercept_)
train_score=reg.score(x_train,y_train)
test_score=reg.score(x_test,y_test)
print('TrainScore(R-Squared):',train_score)
print('TestScore(R-Squared):',test_score)
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
print('Mean Squared Error:',mse)
print('Mean Absolute Error:',mae)
```
